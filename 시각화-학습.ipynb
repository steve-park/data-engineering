## figure

- Matplotlib에서 그래프(Axes, plot)는 Figure 객체내에 존재함
- 그래프를 위한 새로운 Figure는 plt.figure() 함수를 통해 생성
- plt.figure(번호, figsize=크기, dpi=1인치당 픽셀)
    + Create a new figure, or activate an existing figure.
    + cell에는 기본 figure가 설정되어 있음 : 0번
    + 번호를 지정하면 지정한 번호의 figure에 그려짐, 입력된 번호의 figure가 없거나, 생략되면 새로운 figure 생성
    + figsize=(가로크기, 세로크기)로 figure 크기 조절
    + 크기 단위 : 인치, 기본 1인치 = 72 픽셀 (dpi를 사용해 조절 가능)
- Cell : 텍스트를 포함한 사각형 (Cell은 여러 개의 Figure를 포함할 수 있음)
- Figure : 전체 캔버스  (Figure는 여러 개의 Axes를 포함할 수 있음)
- Axes : 1개 그래프 (Axes는 여러 개의 Plot을 포함할 수 있음)
- Plot : 1개 선

## 선 그래프 (Line Plot)

- `plt.plot()` 함수 사용
- `plt.plot([x], y, [fmt], [data])` 함수를 사용하여 x, y 축을 갖는 그래프 생성
    - x, y : array-like or scalar
    - fmt : str, optional, A format string, e.g. 'ro--' for red circles with dashed line
    - 'ro--' : fmtcolor='red', marker='o', linestyle='dashed'
    - data : indexable object, optional, DataFrame 같은 객체를 주면, x, y를 column 이름으로 줄 수 있음
- plot 설명에서 색, marker 등에 대한 정보 확인
- https://matplotlib.org/2.0.2/examples/color/named_colors.html
- https://matplotlib.org/api/_as_gen/matplotlib.pyplot.plot.html

### 하나의 선 그리기

- `plt.plot(y)` : x 는 [0,1,2,...] 자동 완성
- `plt.plot(x, y)` : x,y 둘 다 지정, 갯수가 동일해야 함
- `plt.plot(x, y, fmt)` : 선의 스타일 지정

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
score = pd.Series([1340, 1500, 2400, 3500, 4500, 5000, 5500, 4900, 4200, 3000, 2000, 1500], 
                  index= ['Jan','Feb','Mar','Apr','May','Jun','Jul','Aug','Sep','Oct','Nov','Dec'])
```

```python
plt.plot(score.values)
plt.show()
```

```python
plt.plot(score.index, score.values)
plt.show()
```

```python
# plt.plot(score.index, score.values, 'rs--')
plt.plot(score.index, score.values, color='green', marker='s', markersize=9, linestyle='dashed')
plt.show()
```

### 여러 선 그리기

- 하나의 plot 함수에 여러 선 정보 입력 : `plt.plot(x, y, fmt, x2, y2, fmt2, ...)`
- `plt.plot(x, y, fmt)` 여러 번 사용
    + `plt.plot(X1, y1, fmt1)`
    + `plt.plot(x2, y2, fmt2)`

```python
x = np.arange(1, 11)
y1 = x ** 2
y2 = x ** 3
y3 = x * 4

plt.plot(x, y1, 'ro--', x, y2, 'gs--', x, y3, 'b^--')

plt.show()
```

- `plt.ylabel(str, fontsize)` : y 축에 대한 레이블 표시
- `plt.title(str, fontsize)` : plot에 대한 제목

```python
score = pd.Series([1340, 1500, 2400, 3500, 4500, 5000, 5500, 4900, 4200, 3000, 2000, 1500], 
                  index= ['Jan','Feb','Mar','Apr','May','Jun','Jul','Aug','Sep','Oct','Nov','Dec'])

plt.figure(figsize=(7,5))
plt.plot(score, 'b--')
plt.title('Monthly Sales Volume', fontdict={'fontsize': 15})
plt.xlabel('Month')
plt.ylabel('Sales Volume')
plt.show()
```

### 범례 표시

- Legend(범례) 설정을 위해 그래프에 Label 이 있어야 함
- 방법 1
    + `plt.plot(x, y, style, label='text')`
    + `plt.legend(loc='text')`
- 방법 2
    + plt.plot 에 Label 생략
    + `plt.legend(labels=[label1, label2, ...], loc='text')`
- loc : {upper left, upper center, upper right, center left, center, center right, lower left, lower center, lower right, best, right}

```python
x = np.arange(1, 11)
y1 = x ** 2
y2 = x ** 3
y3 = x * 4

plt.plot(x, y1, 'ro--', label='red')
plt.plot(x, y2, 'gs--', label='green')
plt.plot(x, y3, 'b^--', label='blue')
plt.legend(loc='best')

plt.show()
```

```python
x = np.arange(1, 11)
y1 = x ** 2
y2 = x ** 3
y3 = x * 4

plt.plot(x, y1, 'ro--', x, y2, 'gs--', x, y3, 'b^--')
plt.legend(labels=['red', 'green', 'blue'], loc='upper center')

plt.show()
```

### 눈금선 표시

- `plt.grid()` : 눈금선 표시 (True 사용 가능)
- `plt.grid(False)` : 눈금선 표시 안 함
- `plt.grid(which='both', axis='both', color='gray', dashes=(3,3))`
    + 종류 지정: which={'major', 'minor', 'both'}
    + 축 지정 : axis={'both', 'x', 'y'}
    + 점선 모양 지정 : dashes=(5,2,1,2) describes a sequence of 5 point and 1 point dashes separated by 2 point space
    + 투명도 지정 : alpha=0~1
    + linestyle, linewidth : 눈금선 모양과 두께지정 ('-','--','-.',':' 등)
- https://matplotlib.org/stable/api/_as_gen/matplotlib.pyplot.grid.html

```python
x = np.arange(1, 11)
y1 = x ** 2
y2 = x ** 3
y3 = x * 4

plt.plot(x, y1, 'ro--', x, y2, 'gs--', x, y3, 'b^--')
plt.legend(labels=['red', 'green', 'blue'], loc='best')

# plt.grid(True)
# plt.grid(which='both', axis='y', color='gray', alpha=0.5, dashes=(5,2,1,2))
plt.grid(which='both', axis='y', color='red', alpha=0.5, linewidth=1, linestyle=':')

plt.show()
```

### 눈금 간격 (Tick) 설정

- Ticks : 그래프의 축에 간격을 구분하기 위해 표시하는 눈금
- 방법 1: `ax = plt.axes()` 를 통해 axes 객체를 받아 작업
    + `Axes.set_xticks(self, ticks, *, minor=False)`
    + `Axes.set_yticks(self, ticks, *, minor=False)`
    + 주 눈금 (Major Tick) 설정
        - `ax.set_xticks(array-like)`
        - `ax.set_yticks(array-link)`
    + 보조 눈금 (Minor Tick) 설정
        - `ax.set_xticks(array-like, minor=True)`
        - `ax.set_yticks(array-link, minor=True)`
- 방법 2 : plt.xticks 및 plt.yticks 사용
    + `plt.xticks(ticks=None, labels=None, **kwargs)`, `plt.yticks(ticks=None, labels=None, **kwargs)`
        - ticks : array-like, optional
        - lables : array-like, optional
        - kwargs : Text Property
    + xticks: https://matplotlib.org/stable/api/_as_gen/matplotlib.pyplot.xticks.html
    + kwargs : https://matplotlib.org/stable/api/text_api.html#module-matplotlib.text
    + `plt.tick_params()` : 눈금의 스타일 지정
    + https://matplotlib.org/3.1.1/api/_as_gen/matplotlib.pyplot.tick_params.html

```python
ax = plt.axes()
```

```python
import matplotlib.ticker as ticker

x = np.arange(1, 11)
y1 = x ** 2
y2 = x ** 3

ax = plt.axes()
ax.set_xticks(np.arange(1, 11, 2))
ax.set_yticks(np.arange(0, 1001, 200))
ax.set_xticks(np.arange(1, 11), minor=True)
ax.set_yticks(np.arange(0, 1001, 100), minor=True)

plt.plot(x, y1, 'ro--', x, y2, 'gs--')
plt.grid(which='major', axis='both', color='blue', dashes=(3,3))
plt.grid(which='minor', axis='both', color='gray', dashes=(3,3))

plt.show()
```

### 설명 화살표, 글 표시

- `plt.annotate(text, xy, xytext, arrowprops)` -> <b>매우 다양한 형식이 있음!</b>
    + text : 표시할 문구
    + xy : 화살표를 표시할 좌표
    + xytext : 문구를 표시할 좌표
    + arrowprops : 화살표 속성
- annotations: https://matplotlib.org/stable/tutorials/text/annotations.html#annotations
- `plt.text(x, y, text)`
    + x, y : 문구 표시할 위치 좌표
    + text : 표시할 문구

```python
x = np.arange(1, 11)
y1 = 3 * x ** 2
y2 = 2 * x ** 3

plt.plot(x, y1, 'ro--', x, y2, 'gs--')
# here = 8, 1080 (arrow 위치), 7, 1350 (text 위치)
# 3x^2 = 8.5, 100
# 2x^3 = 8.5, 1100

plt.annotate('here', (8, 1080), (7, 1350), arrowprops={'color': 'blue'})
plt.text(8.5, 100, '$y=3x^2$')
plt.text(8.5, 1100, '$y=2x^3$')

plt.grid()
plt.show()
```

## Subplot

- 여러 plot 을 나누어 표시
- `plt.subplot(row, col, idx)`
    + row, col : shape (행, 열)
    + idx : 일련번호, 1 부터 시작
- `figure, axes = plt.subplots(row, col, **kwargs)`
    + row, col : shape (행, 열)
    + sharex = True : x축을 공유한다는 의미
    + sharey = True : y축을 공유한다는 의미    
    + constrained_layout = True : subplot 간 간격 확보 - 제목 달 떄
- figure : 1개의 캔버스
- axes : 캔버스 위에 그려지는 1개이 그래프
- https://matplotlib.org/stable/api/_as_gen/matplotlib.pyplot.subplot.html
- https://matplotlib.org/stable/api/_as_gen/matplotlib.pyplot.subplots.html

![nn](./images/시각화-02.png)

```python
name = ["Kim", "Yoon", "Choi", "Park"]
month = ["March", "April", "May", "June"]
data = {'name'    : month,
        'english' : [80, 90, 100, 95],
        'chinese' : [100, 80, 70, 85],
        'korean'  : [95, 100, 80, 60],
        'math'    : [90, 70, 60, 100]}
df = pd.DataFrame(data)
df
```

```python
plt.subplot(2,2,1) # plt.subplot(221)
plt.plot('name', 'english', 'ro--', data=df)
plt.ylim(60, 100)
plt.grid()

plt.subplot(2,2,2) # plt.subplot(222)
plt.plot('name', 'chinese', 'gs--', data=df)
plt.ylim(60, 100)
plt.grid()

plt.subplot(2,2,3) # plt.subplot(223)
plt.plot('name', 'korean', 'b^--', data=df)
plt.ylim(60, 100)
# plt.grid()

plt.subplot(2,2,4) # plt.subplot(224)
plt.plot('name', 'math', 'm--', data=df)
plt.ylim(60, 100)
plt.grid()

plt.show()
```

```python
x = 'name'
y = ['english','chinese','korean','math']
fmt = ['ro--','gs--','b^--','m--']
r, c = 2, 2

for number, idx in enumerate(range(r * c), start=221):
    plt.subplot(number)
    plt.plot(x, y[idx], fmt[idx], data=df)
    plt.ylim(60, 100)
    plt.grid(axis='y')

plt.show()
```

```python
names = ['group_a','group_b','group_c']
values = [10, 50, 100]
graphs = [plt.bar, plt.scatter, plt.plot]

plt.figure(0, figsize=(10,3), facecolor='#ACFF44')

for location, idx in enumerate(range(3), start=131):
    plt.subplot(location)
    graphs[idx](names, values)

plt.suptitle('Categorical Plotting')
plt.show()
```

```python
name = ["Kim", "Yoon", "Choi", "Park"]
month = ["March", "April", "May", "June"]
data = {'name'    : month,
        'english' : [80, 90, 100, 95],
        'chinese' : [100, 80, 70, 85],
        'korean'  : [95, 100, 80, 60],
        'math'    : [90, 70, 60, 100]}
df = pd.DataFrame(data)
df
```

```python
x = 'name'
y = ['english','chinese','korean','math']
fmt = ['ro--','gs--','b^--','m--']
r, c = 2, 2

fig, axes = plt.subplots(r, c, sharex=True, sharey=True)

for axflat, cy, style in zip(axes.flat, y, fmt):  #for axflat, cy, style in zip(axes.flat, y[:r*c], fmt[:r*c]):
    axflat.plot(x, cy, style, data=df)
    # 축 (axis) ticks 지정, ticks 를 없앨 때 [] 사용
    axflat.set(xticks=[0,1,2,3], yticks=range(60,101,5))
    # axflat.axis('off') # axis 를 표시하지 않음
    axflat.axis(['March', 'June', 60, 100])
    axflat.grid()

plt.show()
```

## Scatter Plot

- 두 데이터간의 관계를 보기 위해 사용함
- `plt.scatter(x, y, s, c, alpha)`
    + s : scale
    + c : color
    + alpha : 투명도 (0 ~ 1)
- https://matplotlib.org/stable/api/_as_gen/matplotlib.pyplot.scatter.html#matplotlib-pyplot-scatter

```python
np.random.seed(100)

x = np.arange(50)
y = x + np.random.rand(50) * 10
size = np.random.randint(10, 100, 50)
color = np.random.randint(200, 203, 50)

ax = plt.axes()
ax.set_xticks(np.arange(0, 51, 10))
ax.set_yticks(np.arange(-10, 71, 10))
ax.set_xticks(np.arange(0, 51, 5), minor=True)
ax.set_yticks(np.arange(-10, 71, 5), minor=True)

plt.grid(which='both', color='gray', dashes=(3,3))

plt.scatter(x, y, s=size, c=color, alpha=0.7)

plt.show()
```

## 히스토그램 (Histogram)

- 변수가 하나인 단변수 데이터의 빈도수를 그래프로 표현 (일변량 그래프)
- 히스토그램을 그리기 위한 hist 명령은 bins 인수로 데이터를 집계할 구간 정보를 받음
- 데이터 집계 결과를 반환
- `n, bins, patches = plt.hist(x, bins, range)`
    + x : 데이터 
    + bins : 계급 (x축의 값)
    + range : 대상 범위, (min, max)
    + n : 도수 (y축의 값 = 갯수)
    + patches : 계급별 히스토그램 좌표와 높이
- https://matplotlib.org/stable/api/_as_gen/matplotlib.pyplot.hist.html#matplotlib-pyplot-hist

```python
np.random.seed(1)

x = np.random.normal(100, 15, 10000)

# 전체 범위에 대해 (별도 지정 필요없음). 50개 계급으로 나눈 히스토그램 그리기
n, bins, patches = plt.hist(x, 50)

plt.grid()
plt.show()
```

```python
# 상기 예시에서 hist(x, 50) 을 통해 bins 를 50개(=50개의 구간) 으로 지정하였음
# 18~21 구간의 n (도수/y축) 의 값과 해당 구간의 데이터 (x축)를 확인  
print(n[18:22], bins[18:22])
```

```python
# 상기 예시에서 hist(x, 50) 을 통해 bins 를 50개(=50개의 구간) 으로 지정하였음
# 18~21 구간의 중간에 위치한 20번째 구간의 xy 값과 폭, 높이(y축) 확인
print(patches[20])
```

```python
np.random.seed(1)

mu, sigma = 100, 15
x = mu + sigma * np.random.randn(10000)

# facecolor = 색상, 막대의 색상 지정
# density=True, y축을 확률밀도로 표시

plt.hist(x, 50, density=True, facecolor='green')

plt.xlabel('Smarts')
plt.ylabel('Probability')
plt.text(60, 0.025, f'$\mu={mu},\ \sigma={sigma}$')
plt.axis([40, 160, 0, 0.03])
plt.grid(False)

plt.show()
```

## Pie Graph

- 전체에 대한 각 부분의 비율을 부채꼴 모양으로 나타낸 그래프
- 각 부태꼴의 중심각이 전체에서 차지하는 비율을 나타내며, 비율을 한눈에 볼 수 있다는 장점이 있음
- 전체적인 비율을 쉽게 파악할 수 있어서 언론사에서 통계 수치를 공개할 때 자주 활용됨
- `plt.pie(data, labels, colors, autopct)`
    + data : 데이터
    + labels : 각 부채꼴의 레이블
    + colors : 각 부채꼴의 색상
    + autopct : 비율 표시 형식
- https://matplotlib.org/stable/api/_as_gen/matplotlib.pyplot.pie.html#matplotlib-pyplot-pie
