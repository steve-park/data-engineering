# 시계열 데이터 분석

## 시계열 분석 개요

- 시계열 데이터 : 순차적인 시간의 흐름에 따라 기록된 데이터를 의미
- <b><span style='color:red'>Y = T + S + C + R</span></b> 또는 <b><span style='color:red'>Y = T X S X C X R</span></b>
- 추세 (Trend) : 시간의 흐름에 따라 점진적이고 지속적인 변화
- 계절성 (Seasonality) : 특정 주기에 따라 일정한 패턴을 갖는 변화
- 사이클 (Cycle) : 경제 또는 사회적 요인에 의한 변환(예. 경기 변동)이며 <u>일정 주기가 없고 장기적인 변화</u>
- 잔차 (Residuals) : 설명할 수 없는 변화

- 시계열 데이터 분석의 목적 : <b><span style='color:green'>설명할 수 없는 에러 즉, 잔차(Residual, Error, $\epsilon$)를 최대한 줄이고, 백색잡음(White Noise)으로 만드는 것</span></b>

## 시계열 분석의 특징

- 현재 시점의 시계열 데이터를 분석하는데 <b><span style='color:red'>이전 시간의 값이 현재에도 영향을 끼칠 것이라는 가정</span></b>하에 회귀분석을 진행


![](./images/02-1.png)

## 시계열 분석 vs. 단순 회귀

### 시계열 분석
- 자기상관(Autocorrelation) 존재
- 대표적으로 <u>자기회귀, 이동평균, 자기회귀누적이동평균, 벡터자기회귀 모델 등</u>이 존재
- 현재 시점에 가까운 데이터일수록 서로 강한 관계를 맺는 경향 존재

### 단순 회귀
- 자기상과(Autocorrelation) 없음
- 독립변수와 종속변수는 서로 다른 변수일 경우가 많음 (예. 시계열: 어제의 집값으로 오늘의 집값 예측, 단순회귀: 방갯수, 욕실 여부 등으로 집값 예측)
- 선형 회귀로 시계열 데이터를 분석하려면 더 까다로운 가정 필요 -> 선형성 가정이 필요

## 자기회귀 모델(AR)

- AR 모델은 시계열의 미래 값이 과거 값에 기반한다는 모델. 즉, 이전 값의 영향을 받는 것이 특징
- 시계열 이외의 정보가 없을 경우 가장 먼저 적용하는 모델
- AR(p-차수) -> 예. AR(2): 오늘 변수를 설명하기 위해 어제와 그제의 변수값을 독립변수로 사용

![](./images/02-2.png)
- $\alpha$: 상수항, $\epsilon$: 잔차, 에러

## 이동평균 모델(MA)

- <b><span style='color:red'>전체적인 평향성</span></b>을 다루는 모델로, 설명변수가 <b><span style='color:red'>최근 오차항</span></b>으로만 <b><span style='color:red'>구성</span></b>되어 있는 것이 특징
- 모델이 설명하지 못하는 잔차/에러만을 가지고 분석
- 강한 평균 회귀(평균에 수렴하는) 성향이 있음
- MA(q-차수) -> MA(q): 오늘 변수를 설명하기 위해 어제와 그제의 잔차값을 독립변수로 사용

![](./images/02-3.png)

## 자기회귀누적이동평균 모델(ARIMA)

- <b>AR</b>과 <b>MA</b>를 동시에 고려하고, <b>누적(I)</b>으로 추세까지 고려한 모델로, <b><span style='color:red'>'자기회귀 누적 이동평균 모델'</span></b>이라고도 불림
- 누적(I)는 차분을 고려하는 것임
- ARIMA(p,d-차분,q)

![](./images/02-4.png)

## 정상성(Stationary)

- 정상성을 나타내는 시계열은 관측치가 시간과 무관하여야 함. 즉, <u><b>시간에 관계없이 일정한 평균과 분산</b></u>을 갖고 있어야 함
- 정상성을 띄는 시계열은 해당 시계열이 관측된 시간과 무관 (죽, 시간에 따라 상승하거나 주기적인 변화가 있는 추세나 계절성이 없음)
- 특징
  1. 정상 시계열은 평균이 일정
  2. 분산이 시점에 의존하지 않음
  3. 공분산 또한 시점에 의존하지 않음 (시차에는 의존)
- 정상성을 띄는 시계열은 장기적으로 예측 불가능한 시계열 (e.g 백색잡음 White Noise 가 대표적인 예)
- 잔차의 평균이 0 인 것이 정상 시계열???

- 비정상성 시계열은 시간에 영향을 받는 시계열 (추세나 계절성이 있는 것이 대표적인 특징)
- 특징
  1. 시간의 흐름에 따라 시계열의 평균 수준이 다름
  2. 시간의 흐름에 따라 추세를 가짐 (우상향, 우하향 추세 등)
  3. 시간의 흐름에 따라 계절성이 있음
  4. 시간의 흐름에 따라 시계열의 분산이 증가하거나 감소함
- 비정상 시계열 예제) 여름에 아이스크림 판매량이 높으며, 겨율에 판매량이 낮다 (계절성)

![](./images/02-5.png)

- 자기 상관 그래프에서 0 -> corr($y_t$, $y_t$), 1 -> corr($y_t$, $y_{t-1}$), 2 -> corr($y_{t-1}$, $y_{t-2}$), 3 -> corr($y_{t-2}$, $y_{t-3}$),...
- 따라서, 0: corr($y_t$, $y_t$)은 항상 높은 상관계수를 나타냄
- 또한 상기 예시에서 원계열 ACF 그래프는 정상성이 없으나, 차분 ACF 그래프는 정상성을 나타냄

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

import statsmodels.api as sm

from statsmodels.tsa.stattools import adfuller           # ADF 는 정상성 검정을 위해 사용
from statsmodels.tsa.seasonal import seasonal_decompose  # 시계열 요소 분해
from statsmodels.tsa.arima_model import ARIMA            # ARIMA 모델, SARIMA 모델

import pmdarima as pm                                    # Auto ARIMA: 최적의 ARIMA 차수를 정해줌
```

## 정상 시계열 - 백색잡음 (White Noise)

```python
np.random.seed(1)
x = np.random.randn(365)
```

### 윈계열 시도표 (Time Plot)

```python
plt.figure(figsize=(10,3))

plt.plot(np.arange(365), x)
plt.title('White Noise Time Series')
plt.xlabel('periods (t)')

plt.show()
```

### Augmented Dicky Fuller Test (단위근 검정, ADF test)

<b>Augmented Dicky Fuller Test 가설 검정</b>
- $H_0$ : $\theta$ = 0 -> <u>정상성이 있는 시계열이 아님 (단위근)</u>
- $H_1$ : $\theta$ < 0 -> <u>정상성이 있는 시계열</u>
- 귀무가설을 기각해야 정상성이 있는 시계열
- 수식: $\Delta$$y_t$ = $\alpha$ + $\theta$$y_{t-1}$ + $\gamma_t$$\Delta$$y_{t-1}$ + ... +  $\gamma_p$$\Delta$$y_{t-p}$ + $\epsilon_t$

```python
result = adfuller(x)
result
```

```python
print(f'ADF statistics: {result[0]:,.4f}')
print(f'ADF p-value   : {result[1]:,.4f} - {"H0기각(정상시계열)" if result[1] < 0.05 else "H0채택(비정상시계열)"}')
print('Critical Values (유의수준에 따른 검정통계량): ')
for key, value in result[4].items():
    print(f'   {key}  {value:,.4f}')
```

### ACF와 PACF

```python
fig, axes = plt.subplots(1, 2, figsize=(15,4))

fig = sm.graphics.tsa.plot_acf(x, lags=40, ax=axes[0])
fig = sm.graphics.tsa.plot_pacf(x, lags=40, ax=axes[1])
fig.suptitle('ACF and PACF Graph of White Noise', y=1.05)

plt.show()
```

## 비정상 시계열 - 추세 존재

```python
trend = np.linspace(1, 15, 365) # 추세 생성
x_w_trend = x + trend           # 백색잡음 x 에 추세 추가
```

### 원계열 시도표 (Time Plot)

```python
plt.figure(figsize=(10,3))

plt.plot(np.arange(365), x_w_trend)
plt.title('Non-Stationary Time Series with Trends')
plt.xlabel('periods (t)')

plt.show()
```

### Augmented Dicky Fuller Test (단위근 검정, ADF test)

```python
result = adfuller(x_w_trend)
```

```python
print(f'ADF statistics: {result[0]:,.4f}')
print(f'ADF p-value   : {result[1]:,.4f} - {"H0기각(정상시계열)" if result[1] < 0.05 else "H0채택(비정상시계열)"}')
print('Critical Values (유의수준에 따른 검정통계량): ')
for key, value in result[4].items():
    print(f'   {key}  {value:,.4f}')
```

### ACF와 PACF

```python
fig, axes = plt.subplots(1, 2, figsize=(15,4))

fig = sm.graphics.tsa.plot_acf(x_w_trend, lags=40, ax=axes[0])
fig = sm.graphics.tsa.plot_pacf(x_w_trend, lags=40, ax=axes[1])
fig.suptitle('ACF and PACF Graph of Non-Stationary Time Series with Trends', y=1.05)

plt.show()
```

- 자기상관(Autocorrelation)은 시간이 흐를수록 줄어들고 있음 (파란 음영 부분 안으로 들어옴)
- 편자기상관(Partial Autocorrelation)은 시차 5번째부터 파란 음영부분으로 들어왔음
- 상기 예제와 같이 ACF 에서 천천히 파란 음영으로 들어오면서 PACF 는 급격하게 파란 음영으로 들어오는(떨어지는) 경우 : AR 모델 고려

### ARIMA -- AR(5)

```python
arima_mod = ARIMA(x_w_trend, order=(5, 0, 0))  # ARIMA(5,0,0) == AR(5)
```

```python
arima_mod = sm.tsa.arima.ARIMA(x_w_trend, order=(5,0,0))
```

```python
result = arima_mod.fit()
```

```python
result.fittedvalues
```

```python
# 실데이터 vs 모델 결과
plt.figure(figsize=(10,3))
plt.plot(np.arange(365), x_w_trend)
plt.plot(np.arange(365), result.fittedvalues)
plt.show()
```

```python
result.summary()
# Model : ARIMA(5,0,0) 즉, AR(5)모델
# Log Likelihood : 음수가 낮아져야 좋음 모델 즉, 0에 가까울수록 좋음
# AIC, BIC : 값이 작을수록 좋음 (Information Criteria)
# ar.L1 (오늘 변수???), ar.L2(어제 변수???), ar.L3(그제 변수???)
```

- Ljung-Box (L1) (Q) 통계량의 귀무가설 "잔차(residual)는 백색잡음(white nodise)"
- <u>상기 예시에서 Ljung-Box (L1) (Q):	0.54(%)이므로 5% 미만</u>
- Jarque-Bera (JB) 통계량의 귀무가설 "잔차(residual)는 정규성을 만족"
- <u>상기 예시에서  Jarque-Bera (JB):	0.20(%)이므로 5% 미만</u>
- <b>잔차가 정규성을 만족한다면, skew는 0, kurtosis는 3에 가까운 것이 특징임</b>

## 시계열 성분 분해 (Time Series Decomposition)

### 데이터 로드

```python
df = pd.read_csv('./icecream.csv')
```

```python
df.head()
```

```python
df.info()
```

### 데이터 변환

```python
df['time'] = pd.to_datetime(df.year.astype('str') + '/' + df.month + '/' + df.date.astype('str'))
```

```python
df = df.set_index('time', drop=True)
df.drop(columns=['year', 'month', 'date'], inplace=True)
```

```python
df.info()
```

```python
ts = df['sales_amount']
```

### 시도표

```python
ts.plot(figsize=(10,3))
plt.show()
```

시도표를 볼 때 생각해볼 점
1. 추세는 어떠한가? 우상향인가? 우하향인가?
2. 계절성은 어떠한가? 반복되는 패턴은 있는가?
3. 분산은 어떠한가? 시간의 흐름에 상관없이 일정한가? 아니면 증가하는가?

### ACF와 PACF

```python
fig, axes = plt.subplots(1, 2, figsize=(15,4))

fig = sm.graphics.tsa.plot_acf(df, lags=40, ax=axes[0])
fig = sm.graphics.tsa.plot_pacf(df, lags=40, ax=axes[1])
# fig.suptitle('ACF and PACF Graph of Icecream', y=1.05)

plt.show()
```

생각해 볼 점 : 자기상관의 패턴은 어떠한가? 또한 편자기 상관의 패턴은 어떠한가?
- 계절성이 있는 경우 ACF 는 위와 같이 파도치 듯 한 패턴을 가짐. PACF 의 경우에도 어떠한 패턴을 가짐

### 계절성 분해 (Seasonal Decompose)

```python
# additive 분해방식 -> 계절성이 과거와 현재 모두 동일한 경우 사용 즉, 잔차가 등분산성일 떄 사용
# period : 12 -> 1년 주기 계절성
decomp = seasonal_decompose(ts, model='additive', period=12)
```

```python
# 시계열 분해 그래프
fig = decomp.plot()
fig.set_size_inches((12,8))
# Plot 순서
# 1. 원계열 시도표
# 2. Trend : 추세
# 3. Seasonal : 계절성
# 4. Resid : 잔차 -> 하기 예시는 이분산성
```

분해 후 남는 잔차(Residual)를 살표보자! 시간의 흐름에 따라 동일한가? (homoskedastic) <b><u>증가하는가? 감소하는가? (heteroskedastic)</u></b>

```python
# multiplicative 분해방식 -> 잔차가 이분산성일 떄 사용
# period : 12 -> 1년 주기 계절성
decomp = seasonal_decompose(ts, model='multiplicative', period=12)
```

```python
# 시계열 분해 그래프
fig = decomp.plot()
fig.set_size_inches((12,8))
```

분해 후 남는 잔차(Residual)를 살표보자! <b><u>시간의 흐름에 따라 동일한가? (homoskedastic)</u></b> 증가하는가? 감소하는가? (heteroskedastic)

### SARIMAX : ARIMA + 계절성(S) = SARIMA
(참고: SARIMAX의 X는 Exogenous Variable을 의미하며, 우리는 외생변수를 사용하지 않을 것이기에 SARIMA 사용)

```python
# seasonailty 를 ARIMA 모델링
# AR(3) + I(1) + MA(0)
# SAR(3) + SI(1) 계절주기는 12
s_mod = sm.tsa.statespace.SARIMAX(ts, order=(1,1,0), seasonal_order=(1,1,0,12))
```

```python
result = s_mod.fit()
```

```python
result.summary()
```

```python
# 예측 그래프
plt.figure(figsize=(10,3))
plt.plot(ts)
plt.plot(result.fittedvalues)
plt.show()
```

```python
result = s_mod.fit()
```

```python
result.summary()
```

- 원계열 : 추세, 계절성, 싸이클, 잔차로 구성
- 추세 : 시간의 흐름에 따라 점진적인 변화를 관찰
- 계절성 : 계절에 따른 규칙적인 패턴을 확인

## 시계열의 대표 통계 모델

- 자기회귀(AR)
- 이동평균(MA)
- 이동평균을 누적한 자기회귀 모델(ARIMA)
