# 연습 01

- easySample2.csv 파일을 열어, dtype을 올바르게 정리한다
- shelve 바이너리 파일로 저장 (파일이름: mySample)
- 'sample3' key 로 정리된 DataFrame 저장

```python
import pandas as pd
import numpy as np
```

```python
data = pd.read_csv('./data/easySample2.csv', index_col='ID')
data.head()
```

```python
print(data.shape, end='\n\n')
print(data.dtypes, end='\n\n')
print(data.memory_usage(index=True, deep=True), end='\n\n')
```

```python
data['birth'] = pd.to_datetime(data['birth'])

display(data['birth'].dtype)
display(data['birth'])
```

```python
from pandas.api.types import CategoricalDtype

dept_cat = CategoricalDtype(['Accounting', 'Education', 'Marketing', 'Sales'], ordered=False)
data['dept'] = data['dept'].astype(dept_cat)

display(data['dept'].dtype)
display(data['dept'])
```

```python
data['overtime'] = pd.to_timedelta(data['overtime'])

display(data['overtime'].dtype)
display(data['overtime'])
```

```python
a = data.replace(r'^\s+$', np.nan, regex=True)
b = a.replace(r'^(\d+),(\d+)$', r'\1\2', regex=True)
data = b.replace(np.nan, 0)

display(data.head())
```

```python
data = data.astype({'english': 'int32', 
                    'japanese': 'int32', 
                    'chinese': 'int32', 
                    'salary': 'int32'})

display(data.dtypes)
display(data.head())
```

```python
import shelve

with shelve.open('./data/mysample') as myshelf:
    myshelf['sample3'] = data
```

```python
with shelve.open('./data/mysample') as myshelf:
    data2 = myshelf['sample3']

display(data2.head())
display(data2.dtypes)
```

# 연습 - indexing #1

```python
import shelve
import pandas as pd
import numpy as np

myshelf = shelve.open('./data/mysample')
df = myshelf['sample3'][['pname', 'dept', 'english', 'salary']]
df.index = pd.RangeIndex(0, len(df))
df
```

Q) english, salary, dept 컬럼에 대해 single label indexer 를 사용하여 indexing 하여 각각 se, ss, sd 라는 이름을 부여한다.

```python
se = df['english']
ss = df['salary']
sd = df['dept']
```

```python
print(se.head(3), ss.head(3), sd.head(3), sep='\n\n')
```

Q) se가 0보다 큰 것에 대해 boolean indexer 를 사용하여 indexing 한다

```python
df[se > 0]
```

Q) se가 se의 평균값 X 2 보다 큰 것에 대해 boolean indexer를 사용하여 indexing 한다

```python
df[se > (se.mean() * 2)]
```

Q) df에서 ss(salary)가 4500 초과이면서 5000 미만인 것에 대해 boolean indexer 를 사용하여 indexing 한다

```python
df[(ss > 4500) & (ss < 5000)]
```

Q) df에서 ss(salary)가 4500 초과면서 5000 미만인 것에 대해 boolean indexer 를 사용하여 indexing 한다

```python
df[np.logical_and(ss > 4500, ss < 5500)]
```

Q) df에서 sd(dept)가 Marketing 이거나 Accounting 인 것을 boolean indexer 를 사용하여 indexing 한다

```python
df[(sd == 'Marketing') | (sd == 'Accounting')]
```

Q) df에서 sd(dept)가 Marketing 이거나 Accounting 인 것을 boolean indexer 를 사용하여 indexing 한다

```python
df[np.logical_or(sd=='Marketing', sd=='Accounting')]
```

Q)df에서 sd(dept)가 Education이 아닌 것을 boolean indexer 를 사용하여 indexing 한다

```python
df[~(sd == 'Education')]
```

Q)df에서 sd(dept)가 Education이 아닌 것을 boolean indexer 를 사용하여 indexing 한다

```python
df[np.logical_not(sd == 'Education')]
```

# 연습 - indexing #2

```python
import pandas as pd
import numpy as np

name = ['Kim','Yoon','Choi','Park']
data = {'korean': ['A', 'A', 'A', 'B'], 
        'english': ['B', 'A', 'C', 'B']}
```

```python
df = pd.DataFrame(data, index=name)
df.index.name = 'name'
df
```

1. english 컬럼을 s 라는 이름으로 저장한다
2. s에 대해 A 에 해당하면 True, 아니면 False 로 하는 Series 를 생성한다
3. s에 대해 B 또는 C 에 해당하면 True, 아니면 False 로 하는 Series 를 생성한다
4. s1, s2 를 사용하여 df 를 indexing 한다

```python
s = df['english']
s
```

```python
s1 = s.isin(['A'])
s2 = s.isin(['B','C'])
```

```python
display(s1)
display(s2)
```

```python
display(df, df[s1], df[s2])
```

Q) isin()과 동일한 결과를 같도록 비교 연산자를 사용하여 조건을 작성한다

```python
s1 = (s == 'A')
```

```python
s2 = (s == 'B') | (s == 'C')
```

```python
display(df, df[s1], df[s2])
```

Q) df에서 dept가 'Education'이고 chinese가 1인 행에 대해  'pname', 'dept', 'chinese' 열 가져오기

```python
df.loc[(df.dept == 'Education') & (df.chinese == 1),['pname', 'dept', 'chinese']]
```

Q) 19로 시작하면서 dept가 'Marketing'인 행에 대해 'pname', 'dept', 'birth' 열 가져오기

```python
df.loc[np.logical_and((df.index // 1000000) == 19, df.dept == 'Marketing'), 
       ['pname', 'dept', 'birth']]
```

Q) df에서 english 또는 chinese가 1인 행에 대해 'pname', 'english', 'chinese' 열 가져오기

```python
df.loc[np.logical_or(df.english == 1, df.chinese == 1), ['pname','english','chinese']]
```

Q) df에서 overtime이 16시간 이상인 행에 대해 'pname', 'dept', 'overtime' 열 가져오기

```python
df.loc[(df.overtime.dt.seconds // 3600) >= 16,
      ['pname', 'dept', 'overtime']]
```
