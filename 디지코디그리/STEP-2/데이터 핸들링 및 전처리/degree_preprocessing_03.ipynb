### 1. 조건에 맞춰 필터링하기
데이터프레임에서 조건을 충족하는 행을 갖고 올 수 있습니다.<br>
조건식은 ==, >, <, != 등의 기호나 산식을 넣을 수 있습니다.
> df[df['칼럼명'] == 값]

#### 1-1 단일조건 필터링하기
- 데이터프레임에서 성별이 남자인 경우의 데이터를 뽑아보겠습니다. 

#### 🔖  범주형 데이터는 value_counts() 함수를 활용하면 값의 구성을 알 수 있습니다.

```python
# 성별의 구성을 확인합니다. 
df['gender'].value_counts()
```

```python
# 성별이 남자인지를 확인해 봅니다.
df['gender']=='M'
```

```python
# 전체 데이터프레임에서 성별이 남자인 사람으로 필터링한 정보를 가져옵니다.
df[df['gender']=='M']
```

#### 1-2. 복합조건 필터링하기
데이터프레임에서 2가지 이상의 컬럼의 조건, 복합조건을 충족하는 행을 갖고 올 수 있습니다.<br>
복합조건은 하나의 조건이 끝날 때 마다 괄호('(',')')로 구분하여 주어야 하며,<br>
and와 or을 사용하지 않고, **'&', '|'**을 사용해야 합니다.

- 데이터프레임에서 'userId'가 5번인 고객이 2점 이하의 평점을 내린 목록들을 출력해 보겠습니다.

```python
# 전체 데이터프레임에서 'userId'가 5인 고객이 본 영화 목록을 확인합니다.
df[df['userId']==5]

len(df[df['userId']==5])
```

```python
# 전체 데이터프레임에서 고객들이 평점을 2점 이하를 준 영화리스트를 확인합니다.  
df[df['rating']<=2]
```

```python
df[(df['userId']==5) & (df['rating']<=2) ]
```

#### 1-3. 필터링한 결과 중 원하는 정보만 추출하기
조건으로 필터링한 결과에서 다음과 같은 형태로 원하는 정보만 추출할 수도 있습니다.<br>
> df[df['칼럼명'] == 값]['추출할 컬럼명'] 
- 'genres' 컬럼의 정보만 출력해 보겠습니다. 

```python
# 'genres' 만 출력
df[(df['userId']==5) & (df['rating']<=2)]['title']
```

### 2. 컬럼명 변경하기
데이터프레임에 컬럼명을 전체를 변경하거나 일부만 변경할 수 있습니다.


#### 2-1. 전체 컬럼명 변경하기
columns를 사용하여 전체 컬럼명 확인하고 전체 변경할 수 있스니다.<br>
한 개의 컬럼명만 바꾸려고 해도 모든 컬럼을 입력해야 되는 번거로움이 있습니다. <br> 
하나라도 누락되면 오류 메시지가 발생합니다.

- 영화 제목 컬럼 'title'룰 한글 '영화제목' 로 변경해 보세요.  

```python
# 기존 컬럼명 확인하기
df.columns
```

```python
# 새로운 컬럼명 List를 작성하여 대체하기
df.columns = ['movieId', 'imdbId', 'tmdbId', '영화제목', 'genres', 'userId', 'rating',
       'timestamp', 'gender', 'age', 'occupation', 'zipcode']
```

```python
df.head(3)
```

#### 2-2. 선택하여 컬럼명 변경하기
rename 함수의 columns 옵션을 통해 원하는 컬럼의 컬럼명만 변경할 수 있습니다.<br>
columns는 dictionary 형태로 기존 컬럼명과 새 컬럼명으로 맞대응하여 대체할 수 있습니다.
> df.rename(columns={'기존 컬럼명' : '새 컬럼명'})
- 'zipcode' 컬럼의 컬럼명을 '우편번호'로 변경해 보세요.

```python
df = df.rename(columns={'zipcode':'우편번호'})
```

```python
df.head(3)
```

### 3.  새로운 컬럼 생성하기
데이터프레임에 새로운 컬럼을 기존 컬럼 정보의 데이터를 분리하거나 통합하여 새로운 값을 생성할 수 있습니다.

#### 3-1. 새로운 컬럼을 만들고 기본값 세팅하기 
> df['새로운 컬럼명'] = 값 
- 데이터 프레임에 'New' 컬럼을 만들고 기본값으로 0을 세팅 해봅시다.

```python
df['new'] = 0
```

#### 3-2. 인덱싱을 활용하여 원하는 정보를 가져와서 새 컬럼 생성하기 
- '영화제목' 컬럼에서 제작년도를 가져와서 그 정보를 새로운 컬럼 'year'에 저장해 보겠습니다.

#### 🔖 Pandas에서 문자열 관련 함수를 사용하거나 전처리를 하기 위해서는 함수 및 명령어 앞에 str을 붙여주어야 한다.

```python
df.head(3)
```

```python
# df['영화제목']의 정보를 문자열 처리한 후 괄호'(',')'를 제외한 4자리의 숫자만 가져옵니다.
df['영화제목'].str[-5:-1]
```

```python
# 가져온 데이터를 'year' 이라는 새로운 컬럼을 만들고 저장합니다.
df['year'] = df['영화제목'].str[-5:-1]
```

```python
df.head(3)
```

#### 3-3. 기존의 데이터 분리하여 새로운 컬럼 만들기
split() 함수를 사용하면 구분자를 기준으로 데이터를 분리할 수 있습니다.
> str.split('구분자')

- 'genres' 컬럼의 데이터를 구분자 '|' 를 기준으로 분리해 보겠습니다.

```python
# 명령어 앞에 str을 붙이고 split() 함수를 사용합니다. 
df['genres'].str.split('|')
```

##### 📌 expand 옵션을 활용하면, 분리한 List를 바로 새로운 데이터프레임으로 생성할 수도 있습니다. 

```python
# expand=True로 하여 새로운 데이터프레임을 생성합니다.
df['genres'].str.split('|', expand=True).head()
```

### 4. 불 필요한 컬럼 삭제하기

#### 4-1. 특정 컬럼만 삭제하기
pandas의 drop() 함수를 사용하고 columns 옵션을 활용합니다.
> df.drop(columns=['컬럼명1', '컬럼명2])
- 데이터프레임 df에서 'new' 컬럼을 삭제해 보겠습니다.

```python
df.drop(columns=['new'])
```

```python
df.head(3)
```

##### 📌 inplace 옵션을 활용하면, 변경한 내용을 데이터프레임에 바로 적용할 수 있습니다.

```python
# inplace=True로 하여 변경 내용을 바로 적용
df.drop(columns=['new'], inplace=True)
```

```python
df.head()
```

#### 4-2. 필요한 컬럼만 원하는 순서대로 저장하는 방법
데이터를 새로 생성하고, 삭제하다 보면 컬럼의 순서가 뒤죽박죽이 되는 경우가 있습니다.<br>
이런 경우 원하는 순서대로 필요한 컬럼만 뽑아서 다시 저장하는 방법이 있습니다.
- 데이터프레임 df에서 'new' 컬럼은 삭제하고 'year' 컬럼을 '영화제목' 컬럼 다음 순서로 변경해 주세요.

```python
df.columns
```

```python
# df의 컬럼명을 필요한 컬럼과 원하는 순서로 리스트화 합니다.
new_columns = ['movieId', 'imdbId', 'tmdbId', '영화제목', 'year', 'genres', 'userId', 'rating', 'timestamp', 'gender',
       'age', 'occupation', '우편번호']
```

```python
# df에서 new_coluns 리스트의 컬럼들만 뽑아서 다시 저장해 줍니다.
df = df[new_columns]
```

```python
df.head(3)
```
