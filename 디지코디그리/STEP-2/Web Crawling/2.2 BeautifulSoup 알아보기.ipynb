````python
import requests
from bs4 import BeautifulSoup
````

# 1. HTML 문서 가져오기

- HTTPSConnectionPool(host='search.naver.com', port=443) 오류가 발생할 경우 'verify=False'가 추가된 코드 실행

````python

url = 'https://www.aitimes.kr/news/articleList.html?sc_sub_section_code=S2N3&view_type=sm'
# GET 요청하기
# response = requests.get(url)
response = requests.get(url, verify=False)
# 결과 출력
print(response.text)
````

# 2. 가지고 있는 HTML 파싱(구문분석) 하기

````python
soup = BeautifulSoup(response.text, 'html.parser')
````

### 보기 좋게 출력하기

````python
print(soup.prettify())
````

# 3. 원하는 데이터 추출하기

````python
# find_all 사용해서 추출하기
soup.find_all('h4')
````

````python
# select 사용해서 찾기 (css_selector 사용)
soup.select('#section-list li h4')
````

### 제목만 따로 출력해보기

````python
for link in soup.select('#section-list li h4'):
    print(link.text+'\n')
````
