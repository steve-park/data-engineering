# 1. seleniumd에서 webdriver import 하기

````python
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.common.keys import Keys
````

# 2. 웹드라이버 열기

````python
driver = webdriver.Chrome()
````

# 3. 웹페이지로 이동하기

````python
driver.get('https://www.naver.com/')
````

### 웹페이지 HTML 가져오기

````python
print(driver.page_source)
````

# 4. 요소 찾기

````python
search_input = driver.find_element(By.CSS_SELECTOR,"#query")
search_btn = driver.find_element(By.CSS_SELECTOR,"#search-btn")
````

````python
# 요소를 찾는 여러가지 방법이 있어요.

#search_input = driver.find_element(By.ID,"query")
#search_input = driver.find_element(By.CLASS_NAME,"search_input")
````

# 5. 글자 입력하기

````python
search_input.send_keys('KT')
````

````python
# 초기화도 가능해요
search_input.clear()
````

# 6. 요소 클릭하기

````python
search_btn.click()
````
