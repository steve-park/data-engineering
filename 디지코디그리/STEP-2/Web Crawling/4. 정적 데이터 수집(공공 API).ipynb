# 공공 데이터 포탈 API - 대구광역시 맛집

```python
# 요청을 위한 라이브러리 
import requests
# 응답 데이터 타입인 json 을 파싱하기 위한 라이브러리
import json
# 응답 데이터 타입인 xml 을 파싱하기 위한 라이브러리
from bs4 import BeautifulSoup
```

### 요청 URL로 요청보내기

- 만약 아래의 셀 실행 시 오류 발생하면 다음의 urllib3 관련 셀 실행

```python
import urllib3
urllib3.disable_warnings(urllib3.exceptions.InsecureRequestWarning)
```

```python
url = 'https://www.daegufood.go.kr/kor/api/ajax.html'
params = {'mode': 'json', 'addr': '달서구'}
# response = requests.get(url, params=params)
response = requests.get(url, params=params, verify=False)
print(response.status_code)
```

```python
url = 'https://www.daegufood.go.kr/kor/api/tasty.html'
params = {'mode': 'json', 'addr': '달서구'}
# response = requests.get(url, params=params)
response = requests.get(url, params=params, verify=False)
print(response.status_code)
```

### 요청받은 데이터 출력하기

```python
print(response.text)
print(type(response.text))
```

#### 1. key:value 형태의 데이터로 변환

##### (1) json.loads() 함수로 파싱하기

```python
# json 데이터 파싱 하는 방법 -> key:value 형태(딕셔너리)
data = json.loads(response.text)
print(data)
print(type(data))
```

##### (2) json() 메서드 사용하기

```python
data = response.json()
print(data)
print(type(data))
```

- json 형태의 데이터를 읽기좋게 출력하기

```python
# print json data in pretty format
print(json.dumps(data, indent=4, ensure_ascii=False))
```

```python
for i in data['data']:
    print(i['MNU'].replace('<br />', ', '))
```

# 대구광역시 북구 관광코스 정보

```python
# Decoding 인증키 사용
url = 'http://apis.data.go.kr/3450000/daegu/bukgu/tourCourse/viewTourCourse'
params ={'serviceKey' : '본인의 API 키', 
         'pageNo' : '1', 
         'numOfRows' : '10', 
         'resultType' : 'xml' } # json 형식도 가능
# response = requests.get(url, params=params)
response = requests.get(url, params=params, verify=False)
print(response.status_code)
```

- 파라미터값 구분하여 요청시 Decoding 인증키 필요

```python
soup = BeautifulSoup(response.text, 'lxml-xml')
print(soup.prettify())
```

```python
# select와 태그이름을 이용하여 원하는 데이터 추출 가능
soup.select('tourCourseList')
```

```python
# 반복문 tourCourseList 태그 안에 있는 데이터를 추출
for course in soup.select('tourCourseList'):
    print(course.select_one('courseNm').text)
    print(course.select_one('distance').text)
    print(course.select_one('time').text)
    print(course.select_one('tourCourse').text)
    print('---')
    
```
