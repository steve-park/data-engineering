### 객체 저장하기
- shelve 모듈을 사용하여 여러 개의 객체를 dict 처럼 저장하고, 가져오기 할 수 있음
- shelve 모듈을 open할 때 파일의 이름을 주고, dict의 삽입, 갱신, 가져오기와 동일한 방법으로 사용함
- colab 환경 사용시, 생성된 my_data 파일을 다운로드 받아서 다음 사용시 업로드 해야 하는 번거로움 있음

```python
# 지금 사용 중인 DataFrame 저장하기
import pandas as pd
import shelve

df = pd.DataFrame({'A':[1, 2, 3], 'B':[4, 5, 6]})
with shelve.open('my_data') as data:
    data['df_youtube'] = df
```
